```python
import csv
import requests

ids = [220555, 73806, 178882, 73701, 221378, 220640, 220681, 178854, 220608, 220532, 220630, 74478, 204571, 220620, 104552, 204374, 220670, 204436, 160541, 204553, 157130, 133439, 160629, 160553, 220659, 220639, 160510, 204426, 220565, 160592, 88256, 160512, 178937, 204364, 220595, 220579, 209787, 178946, 74356, 204506, 220549, 198783, 204528, 156190, 204494, 220618, 74856, 204410, 220566, 220564, 178989, 178975, 73692, 220550, 220701, 204536, 204515, 7283, 220645, 220632, 206018, 73604, 160518, 220559, 220665, 220642, 66828, 220551, 160674, 220676, 220674, 220623, 160598, 220622, 74270, 220561, 92776, 204560, 204404, 220591, 178997, 220644, 74090, 5718, 5976, 6338, 5973, 4642, 5783, 5953, 5979, 5376, 5736, 5895, 5012, 5936, 6027, 4994, 5732, 178836, 5386, 5895, 5012, 178836, 5386]
itens_por_pagina = 100

db = []

for id in ids:
    url_base = f'https://dadosabertos.camara.leg.br/api/v2/deputados/{id}/discursos'
    parametros = {
        'formato': 'json',
        'itens': itens_por_pagina,
        'idLegislatura': '57,56',
        'ordenarPor': 'dataHoraInicio',
        'ordem': 'ASC'
    }

    pagina = 1
    mais_paginas = True

    while mais_paginas:
        parametros['pagina'] = pagina

        try:
            resposta = requests.get(url_base, parametros)
            resposta.raise_for_status()  # Verificar se ocorreu algum erro na requisição

            try:
                dados_json = resposta.json()
                if 'dados' in dados_json and dados_json['dados']:  # Verificar se a resposta contém dados válidos
                    print(f'Página: {pagina}')

                    for dado in dados_json['dados']:
                        texto = dado['transcricao']
                        data = dado['dataHoraInicio']

                        if 'cotas' in texto:
                            db.append([id, 'COTAS', data, texto])
                        elif 'racismo' in texto:
                            db.append([id, 'RACISMO', data, texto])
                        elif 'ideologia' in texto:
                            db.append([id, 'IDEOLOGIA', data, texto])
                        elif 'maternidade' in texto:
                            db.append([id, 'MATERNIDADE', data, texto])
                        elif 'mulheres negras' in texto:
                            db.append([id, 'MULHERES NEGRAS', data, texto])
                        elif 'gênero' in texto:
                            db.append([id, 'GÊNERO', data, texto])
                        elif 'aborto' in texto:
                            db.append([id, 'ABORTO', data, texto])
                        elif 'trans' in texto:
                            db.append([id, 'TRANS', data, texto])
                        elif 'creche' in texto:
                            db.append([id, 'CRECHE', data, texto])
                        elif 'licença parental' in texto:
                            db.append([id, 'LICENÇA PARENTAL', data, texto])
                        elif 'autonomia financeira' in texto:
                            db.append([id, 'AUTONOMIA FINANCEIRA', data, texto])
                        elif 'licença parental' in texto:
                            db.append([id, 'LICENÇA PARENTAL', data, texto])
                        elif 'igualdade salarial' in texto:
                            db.append([id, 'IGUALDADE SALARIAL', data, texto])
                        elif 'representação feminina' in texto:
                            db.append([id, 'REPRESENTAÇÃO FEMININA', data, texto])
                        elif 'paridade' in texto:
                            db.append([id, 'PARIDADE', data, texto])
                        elif 'punitivismo' in texto:
                            db.append([id, 'PUNITIVISMO', data, texto])
                        elif 'Lei Maria da Penha' in texto:
                            db.append([id, 'LEI MARIA DA PENHA', data, texto])
                        elif 'violência sexual' in texto:
                            db.append([id, 'VIOLÊNCIA SEXUAL', data, texto])
                    pagina += 1
                    mais_paginas = 'next' in dados_json['links'][1]['rel']
                else:
                    mais_paginas = False

            except ValueError as e:
                print(f'Erro ao processar os dados JSON: {str(e)}')
                break

        except requests.exceptions.RequestException as e:
            print(f'Erro na requisição à API: {str(e)}')
            break

# Escrever os dados em um arquivo CSV
with open('discursos.csv', 'w', newline='', encoding='utf-8') as arquivo_csv:
    writer = csv.writer(arquivo_csv)
    writer.writerow(['ID', 'Tipo', 'Data', 'Texto'])  # Escrever o cabeçalho

    for linha in db:
        writer.writerow(linha)
```
